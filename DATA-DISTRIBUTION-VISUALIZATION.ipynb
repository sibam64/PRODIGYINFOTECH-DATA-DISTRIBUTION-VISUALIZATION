```python
# ==============================
# World Bank Population Histogram
# ==============================

import pandas as pd
import matplotlib.pyplot as plt

# Load ACTUAL population dataset (NOT metadata file)
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv", skiprows=4)

print("First 5 Rows:")
print(df.head())

print("\nColumns:")
print(df.columns)

# -----------------------------------
# Population data is stored by YEAR
# Example: '2020' column
# -----------------------------------

# Select one year (example: 2020)
year = "2020"

# Convert to numeric
df[year] = pd.to_numeric(df[year], errors="coerce")

# Remove missing values
df = df.dropna(subset=[year])

# -----------------------------------
# Create Histogram
# -----------------------------------

plt.figure(figsize=(8,5))

plt.hist(df[year], bins=15)

plt.title(f"World Population Distribution ({year})")
plt.xlabel("Population")
plt.ylabel("Number of Countries")

plt.show()

# -----------------------------------
# Better Version: Log Scale
# -----------------------------------

plt.figure(figsize=(8,5))

plt.hist(df[year], bins=15)
plt.xscale("log")

plt.title(f"World Population Distribution ({year}) - Log Scale")
plt.xlabel("Population (Log Scale)")
plt.ylabel("Number of Countries")

plt.show()
```
